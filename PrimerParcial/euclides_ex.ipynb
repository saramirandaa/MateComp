```python
#Se inicializan las variables para las cuales queremos saber el MCD
a = 56
b = 98
```

```python
# c, d serÃ¡n igual al valos absoluto de a y b respectivamente
c, d = abs(a), abs(b)
# AquÃ­ se inicializan los valores que representan los coeficientes ð’” y ð’• 
# correspondiente al MCD pero de la primera iteraciÃ³n
c1, d1 = 1, 0
# Segunda iteracaiÃ³n
c2, d2 = 0, 1
```

```python
# mientras de no sea cero, comienza nuestro ciclo
# dentro del ciclo Se calcula el cociente ð’’ y el residuo ð’“ de la divisiÃ³n de ð’„ entre ð’…. 
# DespuÃ©s de ello, se calculan los nuevos valores de ð’“ðŸ y ð’“ðŸ 
# restando ð’’ âˆ— ð’…ðŸ y ð’’ âˆ— ð’…ðŸa ð’„ðŸ yð’„ðŸ, respectivamente.
while d != 0:
    q, r = divmod(c, d)
    c, d = d, r
    c1, c2 = c2, c1 - q * c2
    d1, d2 = d2, d1 - q * d2
```

```python
# Se calculan los coeficientes ð’” y ð’• .ð’” = ð’„ðŸ/( ð’”ð’ˆð’(ð’‚) âˆ— ð’”ð’ˆð’(ð’„)), donde ð’”ð’ˆð’(ð’‚) es la funciÃ³n que devuelve el signo 
# de a (1 si a es positivo, -1 si es negativo o 0 si es cero) y ð’• = ð’„ðŸ/ (ð’”ð’ˆ ð’ ð’ƒ âˆ— ð’”ð’ˆ ð’ ð’„ )

def sgn(x):
    return (x > 0) - (x < 0)

s = c1 / (sgn(a) * sgn(c))
t = d1 / (sgn(b) * sgn(c))
print ("s:", s," t:", t)
print ("MCD:", c)
```
